Read the RDF file

```
from rdflib import Graph, Namespace, Literal
from rdflib.namespace import RDF, RDFS
g = Graph()
g.namespace_manager.bind('ns', Namespace("http://somewhere#"), override=False)
g.namespace_manager.bind('vcard', Namespace("http://www.w3.org/2001/vcard-rdf/3.0#"), override=False)
g.parse(github_storage+"/resources/example5.rdf", format="xml")
```

Create a new class named Researcher

```
ns = Namespace("http://somewhere#")
g.add((ns.Researcher, RDF.type, RDFS.Class))
for s, p, o in g:
  print(s,p,o)
```

**TASK 6.1: Create a new class named "University"**

```
g.add((ns.University, RDF.type, RDFS.Class))
```

**TASK 6.2: Add "Researcher" as a subclass of "Person"**

```
g.add((ns.Researcher, RDFS.subClassOf, ns.Person))
```

**TASK 6.3: Create a new individual of Researcher named "Jane Smith"**

```
janeURI = ns.JaneSmith
g.add((janeURI, RDF.type, ns.Researcher))
```

**TASK 6.4: Add to the individual JaneSmith the fullName, given and family names**

```
VCARD = Namespace("http://www.w3.org/2001/vcard-rdf/3.0#")
fullName = Literal("Jane Smith")
fn = (janeURI, VCARD.FN, fullName)
g.add(fn)

given = Literal("Jane")
givenName = (janeURI, VCARD.Given, given)
g.add(givenName)

family = Literal("Smith")
familyName = (janeURI, VCARD.Family, family)
g.add(familyName)
```

**TASK 6.5: Add UPM as the university where John Smith works**

```
upmURI = ns.UPM
g.add((upmURI, RDF.type, ns.University))
g.add((ns.JohnSmith, VCARD.org, upmURI))
for s, p, o in g:
  print(s,p,o)
```
